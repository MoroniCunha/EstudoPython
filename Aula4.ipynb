```python
# Exercicio 4
# ---Inicio das Variaveis Globais
lista_func = []
cod_func = 0

# ---Fim das Variaveis Globais

# ---Inicio de cadastrar Funcionario
def cadastrar_func(codigo):
  print('Menu de Cadastro de Funcionários')
  print('ID do funcionário: {}' .format(codigo))
  nome = input('Digite o nome do funcionário:')
  setor = input('Digite o setor do funcionário:')
  salario = int(input('Digite o salário do funcionário:'))
  dicionario_func = {'ID'       : codigo,
                     'nome'     : nome,
                     'setor'    : setor,
                     'salario'  : salario}
  lista_func.append(dicionario_func.copy()) # O dicionario servirá para encontrarmos os dados salvos dos funcionários

# --- Fim de cadastrar funcionario

# ---Inicio de consultar Funcionario
def consultar_func():
  print('Menu de Consulta de Funcionários')
  while True:
    opcao_consulta = input('Escolha a opção desejada: \n'+
                          '1 - Consultar todos os Funcionários \n'+
                          '2 - Consultar Funcionário por ID \n'+
                          '3 - Consultar Funcionário por Setor \n'+
                          '4 - Retornar \n'+
                          '>>:')
    if opcao_consulta == '1':
      print('Consulta todos os funcionários')
      for func in lista_func: # func vai varrer toda a lista de funcionários
        print('-------------------------------------')
        for key, value in func.items(): # Varrer todos os conjuntos chave e valor do dicionario func
          print('{}: {}' .format(key, value))
    elif opcao_consulta == '2':
      print('Consulta ID funcionários')
      id_desejada = int(input('Entre com a ID desejada:'))
      for func in lista_func:
        if func['ID'] == id_desejada:
          print('-------------------------------------')
          for key, value in func.items(): # Varrer todos os conjuntos chave e valor do dicionario func
            print('{}: {}' .format(key, value))
    elif opcao_consulta == '3':
      print('Consulta funcionários por Setor')
      setor_desejado = input('Entre com o setor desejado:')
      for func in lista_func:
        if func['setor'] == setor_desejado:
          print('-------------------------------------')
          for key, value in func.items(): # Varrer todos os conjuntos chave e valor do dicionario func
            print('{}: {}' .format(key, value))
    elif opcao_consulta == '4':
      return # Sai dessa opção e retorna ao Main
    else:
      print('Opção Invalida')
      continue

# --- Fim de consultar funcionario

# ---Inicio de remover Funcionario
def remover_func():
  print('Menu de Remoção de Funcionários')
  id_desejada = int(input('Entre com a ID que deseja remover:'))
  for func in lista_func:
    if func['ID'] == id_desejada:
      lista_func.remove(func)
      print('Funcionario removido')
# --- Fim de remover funcionario

# ---Inicio de Main
print('Bem Vindo ao Sistema de Controle de Funcionários do Morôni Cunha Silva')
while True:
  opcao_principal = input('Escolha a opção desejada: \n'+
                          '1 - Cadastrar Funcionário \n'+
                          '2 - Consultar Funcionário \n'+
                          '3 - Remover Funcionário \n'+
                          '4 - Sair \n'+
                          '>>:')
  if opcao_principal == '1':
    cod_func = cod_func + 1
    cadastrar_func(cod_func)
  elif opcao_principal == '2':
    consultar_func()
  elif opcao_principal == '3':
    remover_func()
  elif opcao_principal == '4':
    print('Fim do programa RU 4226094')
    break # Encerra o laço, e termina o programa
  else:
    print('Opção Invalida')
    continue # Volta para o Inicio do laço
# ---Fim de Main
```
